## Exploring the dataset

```python
# Treating missing values in the dataset
# Checking missing values
combined_data.isnull().sum()
```

```python
# Checking unique values in dataset
columns = list(combined_data.columns)
columns.remove('Date')
columns.remove('CompetitionDistance')
```

```python
for col in columns:
    print(col, "  :", combined_data[col].unique())
```

```python
# Extracting year and month from Date
from datetime import datetime

combined_data['Year'] = combined_data['Date'].apply(lambda x: datetime.strptime(x, '%d/%m/%y').year)
combined_data['Month'] = combined_data['Date'].apply(lambda x: datetime.strptime(x, '%d/%m/%y').month)
```

```python
combined_data['Month']
```

```python
# Yearly sales

sns.barplot(x = 'Year',
            y = 'Sales',
            data = combined_data,
            hue = 'Year').set(title = 'Year vs Sales')
plt.legend().set_visible(False)
plt.show()
```

- We don't see significant change on sales over the three years 

```python
# Weekly Sales

sns.barplot(x = 'DayOfWeek',
            y = 'Sales',
            data = combined_data,
            hue = 'DayOfWeek').set(title = 'Week vs Sales')
plt.legend().set_visible(False)
plt.show()
```

- We see the decline from day 1 to 6, but for 7th day, which is Sunday, we see a significant decline.

```python
# Promo-based Sales

sns.barplot(x = 'Promo',
            y = 'Sales',
            data = combined_data).set(title = 'Promo vs Sales')
plt.show()
```
